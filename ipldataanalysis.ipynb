```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the data

ipl=pd.read_csv('matches.csv')
```

```python
#Show the first 5 table from my ipl data sets.....

ipl.head()
```

```python
ipl.shape
```

```python
#Value counts of 'player_of_match' in ipl history....

ipl['player_of_match'].value_counts()
```

```python
#value counts of 'toss winner' in ipl

ipl['toss_winner'].value_counts()[:15]
```

```python
#Show the graph , in which player name with won 'man of the match' ...

plt.figure(figsize=(20,8))
plt.bar(list(ipl['player_of_match'].value_counts()[0:15].keys()),list(ipl['player_of_match'].value_counts()[0:15]), color='b')
plt.xlabel('player_name')
plt.ylabel('player_of_the_match')

plt.show()
```

```python
# count the result of the ipl matches

ipl["result"].value_counts()
```

```python
#values count for how many times have super over or not .....

ipl["super_over"].value_counts()
```

```python
#Graph for team's win more tosses

plt.figure(figsize=(38,15))
plt.bar(list(ipl['toss_winner'].value_counts()[0:15].keys()),list(ipl['toss_winner'].value_counts()[0:15]), color='y')
plt.xlabel('..............team name..............')
plt.ylabel('..............winner times...........')
plt.grid('grid')
plt.show()
```

```python
#Value counts in which 'city' have more ipl matches

ipl['city'].value_counts()[0:15]
```

```python
#Value Counts where was most matches venue

ipl['venue'].value_counts()
```

```python
#Value count  in which seasons had more matches
ipl['season'].value_counts()
```

```python
#Value counts which teams more prefer in both field first or bat first
ipl['toss_decision'].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.bar(list(ipl['toss_decision'].value_counts().keys()),list(ipl['toss_decision'].value_counts()), color="g")
plt.xlabel("........Decision...........")
plt.ylabel("........Count..............")
plt.show()
```
